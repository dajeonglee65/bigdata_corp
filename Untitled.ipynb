```python
익명취재원의 수 / 미확인 서술어 / 투명 취재원
1. 미확인 서술어는 1개 이상을 1점 차감한다.
2. 익명 취재원은 1명 이상을 1점 차감한다.
3. 투명 취재원/익명 취재원 이 1보다 작을경우 1점 차감한다.
4. 비문일 경우 1점 차감한다.
```

```python
df['acc'] = 5
df['a_source'] = df['a_source'].astype('float')

for i in range(df.shape[0]):
    if df['nf_verb'][i] > 0:
        df['acc'][i] -= 1
    
    if df['a_source'][i] >= 1:
        df['acc'][i] -= 1
    
    if df['a_source'][i] == 0:
        df['a_source'][i] = 0.01
        if df['c_source'][i] / df['a_source'][i] < 1:
            df['acc'][i] -= 1
            
    elif df['a_source'][i] != 0:
        if df['c_source'][i] / df['a_source'][i] < 1:
            df['acc'][i] -= 1
    
    if df['non_text'][i] == 1:
        df['acc'][i] -= 1

df.head()
```
