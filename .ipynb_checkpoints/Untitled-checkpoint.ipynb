```python
df['trust']  = 5

for i in range(df.shape[0]):
    if df['source'][i] < 4:
        df['trust'][i] -= 1
    
    if df['email'][i] < 2:
        df['trust'][i] -=1
    
    if df['ratio_sub'][i] > 0.5:
        df['trust'][i] -= 1
        
    if df['non_text'][i] == 1:
        df['trust'][i] = 1

df.head(10)
```
